### __Let add president column in our dataframe__

```python
# Extract the year from the date column
dollar_shilling_exchange['year'] = dollar_shilling_exchange['Date'].dt.year

# Define the function to map years to presidents
def get_president(year):
    if 1978 <= year <= 2002:
        return 'Daniel arap Moi'
    elif 2003 <= year <= 2013:
        return 'Mwai Kibaki'
    elif 2014 <= year <= 2022:
        return 'Uhuru Kenyatta'
    elif year >= 2023:
        return 'William Ruto'
    else:
        return None

# Apply the function to the DataFrame to create a new column for presidents
dollar_shilling_exchange['president'] = dollar_shilling_exchange['year'].apply(get_president)

# Drop the 'year' column if it's not needed
dollar_shilling_exchange= dollar_shilling_exchange.drop(columns=['year'])
```
