### __For NaNs in Stadium Column we fill with the Most used stadium in that country__

#### __First we determine the most frequent stadium for each home team since games are always played in HomeTeam__

```python
from collections import Counter
```

```python
def most_frequent_stadium(group):
    stadium_counts = Counter(group.dropna())
    if stadium_counts:
        return stadium_counts.most_common(1)[0][0]
    return None
```

#### __Next we create a dictionary mapping each home team to its most frequent stadium__

```python
most_frequent_stadiums = afcon1_df.groupby('HomeTeam')['Stadium'].apply(most_frequent_stadium).to_dict()
```

 #### __Here we print the dictionary for debugging__

```python
most_frequent_stadiums
```

####  __we run the function to fill NaN values with the most frequent stadium__

```python
def fill_na_stadium(row):
    if pd.isna(row['Stadium']):
        return most_frequent_stadiums.get(row['HomeTeam'], row['Stadium'])
    return row['Stadium']
```

#### __We apply the function to fill NaN values__

```python
afcon1_df['Stadium'] = afcon1_df.apply(fill_na_stadium, axis=1)
```

#### __We print the resulting for afcon_df__

```python
afcon1_df
```

#### __We do the same case to Cities ,Since all games are played in the cities of the HomeTeam,we shall replace the NaNs in City column with the most used cities of the Home Team__

#### __First we determine the most frequent city for each home team since games are always played in HomeTeam__

```python
def most_frequent_city(group):
    city_counts = Counter(group.dropna())
    if city_counts:
        return city_counts.most_common(1)[0][0]
    return None
```

#### __Next we create a dictionary mapping each home team to its most frequent city__


```python
most_frequent_cities = afcon1_df.groupby('HomeTeam')['City'].apply(most_frequent_city).to_dict()
```

#### __Here we print the dictionary for debugging__

```python
most_frequent_cities
```

#### __we run the function to fill NaN values with the most frequent city__

```python
def fill_na_city(row):
    if pd.isna(row['City']):
        return most_frequent_cities.get(row['HomeTeam'], row['City'])
    return row['City']
```

#### __We apply the function to fill NaN values__

```python
afcon1_df['City'] = afcon1_df.apply(fill_na_city,axis=1)
```

#### __We print the resulting for afcon_df__

```python
afcon1_df
```
